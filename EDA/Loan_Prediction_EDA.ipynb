## Missing Value Plot
**Top portion** of the plot shows the aggregate for each column. Summary statistics is displayed on the right most side.

**Bottom portion** of the plot shows the missing values (black colors) in the DataFrame.

```
import klib

klib.missingval_plot(df)
```

```
non_object_var = ['Gender', 'Dependents', 'Self_Employed', 'ApplicantIncome',
                  'LoanAmount', 'Loan_Amount_Term', 'Credit_History', 'Married']

def fill_with_mode(df,x):
  df[x].fillna(df[x].mode()[0], inplace=True)

for obj in non_object_var:
  fill_with_mode(df, obj)

df.isnull().sum()
```
