```python
#Feature selection
```

```python
data = df_review[["business_id", "user_id", "stars", "text"]]
```

```python
import string
from nltk.corpus import stopwords
from nltk.tokenize import WordPunctTokenizer

stop = []
for word in stopwords.words("english"):
    s = [i for i in word if i not in string.punctuation]
    stop.append(''.join(s))
```

```python
def text_process(mess):
    nopunc = [i for i in mess if i not in string.punctuation]
    nopunc = ''.join(nopunc)
    
    return " ".join([i for i in nopunc.split() if word.lower() not in stop])
```

```python
# Text Cleaning
```

```python
data["text"] = data["text"].apply(text_process)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
user = data[['user_id', 'text']]
business = data[['business_id', 'text']]
```

```python
user = user.groupby('user_id').agg({'text': ' '.join})
business = business.groupby('business_id').agg({'text': ' '.join})
```

```python
# User Tfidf Vectorizer with 5000 Features
```

```python
user_vectorizer = TfidfVectorizer(tokenizer = WordPunctTokenizer().tokenize, max_features=5000)
userid_vectors = user_vectorizer.fit_transform(user['text'])
userid_vectors.shape
```

```python
business_vectorizer = TfidfVectorizer(tokenizer = WordPunctTokenizer().tokenize, max_features=5000)
businessid_vectors = business_vectorizer.fit_transform(business['text'])
businessid_vectors.shape
```

```python
#Matrix Factorization
```

```python
user_matrix = pd.pivot_table(data, values='stars', index=['user_id'], columns=['business_id'])
user_matrix.shape
```

```python
P = pd.DataFrame(userid_vectors.toarray(), index=user.index, columns=user_vectorizer.get_feature_names())
```

```python
Q = pd.DataFrame(businessid_vectors.toarray(), index=business.index, columns=business_vectorizer.get_feature_names())
```

```python
# Gradient Decent Optimization
```

```python
def matrix_factorization(R, P, Q, steps=25, gamma=0.001,lamda=0.02):
    for step in range(steps):
        for i in R.index:
            for j in R.columns:
                if R.loc[i,j]>0:
                    eij=R.loc[i,j]-np.dot(P.loc[i],Q.loc[j])
                    P.loc[i]=P.loc[i]+gamma*(eij*Q.loc[j]-lamda*P.loc[i])
                    Q.loc[j]=Q.loc[j]+gamma*(eij*P.loc[i]-lamda*Q.loc[j])
        e=0
        for i in R.index:
            for j in R.columns:
                if R.loc[i,j]>0:
                    e= e + pow(R.loc[i,j]-np.dot(P.loc[i],Q.loc[j]),2)+lamda*(pow(np.linalg.norm(P.loc[i]),2)+pow(np.linalg.norm(Q.loc[j]),2))
        if e<0.001:
            break
        
    return P,Q
```

```python
P ,Q = matrix_factorization(user_matrix, P, Q, steps=25, gamma=0.001, lamda=0.02)
```

```python
words = "beautiful views at morning bunch"
test_df= pd.DataFrame([words], columns=['text'])
test_df['text'] = test_df['text'].apply(text_process)
test_vectors = user_vectorizer.transform(test_df['text'])
test_v_df = pd.DataFrame(test_vectors.toarray(), index=test_df.index, columns=user_vectorizer.get_feature_names())
```

```python
predictItemRating=pd.DataFrame(np.dot(test_v_df.loc[0],Q.T),index=Q.index,columns=['Rating'])
topRecommendations=pd.DataFrame.sort_values(predictItemRating,['Rating'],ascending=[0])[:7]
```

```python
for i in topRecommendations.index:
    print(df_business[df_business['business_id']==i]['name'].iloc[0])
    print(df_business[df_business['business_id']==i]['categories'].iloc[0])
    print(str(df_business[df_business['business_id']==i]['stars'].iloc[0])+ ' '+str(df_business[df_business['business_id']==i]['review_count'].iloc[0]))
    print('')
```
